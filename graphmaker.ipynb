```python
import matplotlib.pyplot as plt

pwm = []
a0 = []

with open("data.txt") as f:
    for line in f:
        parts = line.strip().split()
        if len(parts) != 2:
            continue
        try:
            p = int(parts[0].split(":")[1])
            r = int(parts[1].split(":")[1])
            pwm.append(p)
            a0.append(r)
        except:
            pass

plt.plot(pwm, label="PWM")
plt.plot(a0, label="A0")
plt.legend()
plt.title("A0 Health Test (10s Sweep)")
plt.show()
```
